```python
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
a_tensor = torch.rand(7, 7)
print(a_tensor)

torch.manual_seed(RANDOM_SEED)
b_tensor = torch.rand(1, 7)
print(b_tensor)

torch.mm(a_tensor, b_tensor.T)
```

```python
torch.cuda.manual_seed_all(1234)
print(torch.cuda.initial_seed())
```

```python
torch.manual_seed(42)
a_tensor = torch.rand(2, 3)
torch.manual_seed(42)
b_tensor = torch.rand(2, 3)
device = "cuda" if torch.cuda.is_available() else "cpu"
a_tensor = a_tensor.to(device)
b_tensor = b_tensor.to(device)
print(a_tensor, "\n", b_tensor)
```

```python
result_mul = torch.mm(a_tensor, b_tensor.T)
result_mul, result_mul.shape, result_mul.device
```

```python
torch.max(result_mul), torch.min(result_mul)
```

```python
torch.argmax(result_mul), torch.argmin(result_mul)
```
